```python
# Assuming 'elements' is the list returned by the partition function
# elements = partition(filename=filename, strategy="hi_res") # Your previous call

Nshow = 50
truncate_length = 1000
print(f"Inspecting the first {min(Nshow, len(elements))} elements out of {len(elements)} total elements found.\n")

for i, element in enumerate(elements[:Nshow]): # Inspect the first Nshow elements
    print(f"--- Element {i+1} ---")
    print(f"Element ID: {element.id}")
    print(f"Type: {type(element).__name__}") # e.g., Title, NarrativeText, Table, Image
    print(f"Category: {element.category}") # Unstructured's classification

    # Print the text content (truncated for brevity)
    if hasattr(element, 'text'):
        element_text = getattr(element, 'text', '') # Use getattr for safety
        if element_text is None:
            print ("No text found.")
        else:
            print(f"Text: {element_text[:truncate_length]}..." if len(element_text) > truncate_length else f"Text: {element_text}")

    # Print all metadata associated with the element
    print("\n  Metadata:")
    if hasattr(element, 'metadata'):
        # Convert metadata to dict if it's an object (like ElementMetadata)
        metadata_dict = element.metadata.to_dict() if hasattr(element.metadata, 'to_dict') else vars(element.metadata)

        if metadata_dict:
            for key, value in metadata_dict.items():
                # Truncate long metadata values like text_as_html or coordinates points
                value_repr = repr(value)
                if len(value_repr) > truncate_length:
                     value_repr = value_repr[:truncate_length] + "..."
                print(f"    * {key}: {value_repr}")
        else:
            print("    * (No metadata found)")
    else:
        print("    * (No metadata attribute found)")

    print("-" * 25 + "\n") # Separator

if len(elements) > Nshow:
    print(f"... and {len(elements) - Nshow} more elements not shown.")
```
